```python
import torch, pickle, time, os, random
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as sci
```

```python
case='test_z2.0_smass'

data=pickle.load(open(f'../../../../scratch/gpfs/cj1223/GraphStorage/{case}/data.pkl', 'rb'))
```

```python
Mhf=[]
Ms=[]
ls=[]
for d in data:
    Mhf.append(d.x.numpy()[0,3])
    Ms.append(d.y.numpy())
    ls.append(len(d.x.numpy()))
Ms=np.hstack(Ms)
Mhf=np.array(Mhf)
ls=np.log10(ls)
```

```python
fig , ax = plt.subplots(3, figsize=(6,12))
ax[0].plot(Mhf,Ms, 'ro',label =' M_halo-M_star')
ax[1].plot(ls,Ms, 'bo',label = 'Tree length-M_star')
ax[2].plot(ls,Mhf, 'ko',label = 'Tree length-M_halo')

for a in ax:
    a.legend()
```

## First, halo spline relationship

```python
#trick to make splines work
Mhf+=np.arange(len(Mhf))/1e6
ls+=np.arange(len(Mhf))/1e6
```

```python
fig , ax = plt.subplots(2, figsize=(6,8))

a=np.array([Mhf,Ms, ls])
a = a[:,a[0,:].argsort()]
Mhf, Ms, ls = a
Mhf+=np.arange(len(Mhf))/1e6
ls+=np.arange(len(Mhf))/1e6

splineMhf=sci.UnivariateSpline(Mhf, Ms)
ax[0].plot(Mhf,Ms, 'ro',label =' M_halo-M_star data')
ax[0].plot(Mhf,splineMhf(Mhf), 'b-',label =' M_halo-M_star spline')
ax[0].set(title=f'sigma ={np.round(np.std(splineMhf(Mhf)-Ms),2)}', xlabel='Mh', ylabel='M_*')

Mhf-=np.arange(len(Mhf))/1e6
ls-=np.arange(len(Mhf))/1e6

a=np.array([Mhf,Ms, ls])
a = a[:,a[2,:].argsort()]
Mhf, Ms, ls = a
Mhf+=np.arange(len(Mhf))/1e6
ls+=np.arange(len(Mhf))/1e6

splinels=sci.UnivariateSpline(ls, Ms)
ax[1].plot(ls,Ms, 'bo',label = 'Tree length-M_star data')
ax[1].plot(ls,splinels(ls), 'r-',label = 'Tree length-M_star spline')
ax[1].set(title=f'sigma ={np.round(np.std(splinels(ls)-Ms),2)}', xlabel='log Tree length', ylabel='M_*')

for a in ax:
    a.legend()
fig.tight_layout()
Mhf-=np.arange(len(Mhf))/1e6
ls-=np.arange(len(Mhf))/1e6
```

## Abundance matching

```python
a=np.array([Mhf,Ms, ls])
a = a[:,a[0,:].argsort()]
Mhf, Ms, ls = a
```

```python
Mhf_am,Ms_am=np.sort([Mhf,Ms], axis=1)
# plt.plot()[0],np.sort([Mhf,Ms])[1])
```

```python
plt.plot(Mhf,Ms,'ro')
plt.plot(Mhf_am,Ms_am,'bo')
plt.title(f'Abundance matched relationship, sigma={np.std(Ms-Ms_am):.2f}')
plt.xlabel('M_halo')
plt.xlabel('M_*')
```
