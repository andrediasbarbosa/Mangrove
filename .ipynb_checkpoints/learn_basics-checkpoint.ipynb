### Trying with pandas

```python
path='../../../../tigress/mcranmer/merger_trees/isotrees/'
pda=pd.read_table(path+'isotree_0_0_0.dat', skiprows=0, delimiter=' ', nrows=10)
```

```python
pd1=pd.read_table(path+'isotree_0_0_0.dat', skiprows=0, delimiter='\s+')
pd2=pd1.drop(axis=0, index=np.arange(50)).reset_index()
```

```python
pd1.drop(axis=0, index=np.arange(50)).reset_index()
```

```python
trees=pd2[pd2.isna()['A[z]']] #these seem to be indices for the trees, not the nodes. 
trees=trees.astype({'id(1)':'int64'}, copy=1)
trees.dtypes
```

```python
halos=pd2[~pd2.isna()['A[z]']] 
```

```python
f,i="float64", "int64"
cs=[f,i,f,i,i,i,i,i,i,f,f,f,f,f,i,f,f,f,f,f,f,f,f,f,f,f]
dicts = {}
keys = halos.columns[1:26]
castto = cs
for i, key in enumerate(keys):
        dicts[key] = castto[i]
print(dicts)

halos=halos.astype(dicts)
```

```python
for c, h, i in zip(halos.columns, halos.dtypes, np.arange(len(halos.dtypes))):
    print(c,h,i)
```

```python
split=np.split(np.array(halos)[:,1:10], np.array(trees.iloc[1:].index)-np.arange(1,len(trees.index)))
```

```python
de, pr=[], []
for k in tqdm(range(len(split))):
    des, pro=[], []
    for i, desc in enumerate(split[k][:,3]):
        if desc in split[k][:i,1]:
            des.append(np.array(desc)
            pro.append(split[k][i,1])
    de.append(des)
    pr.append(pro)
```

```python
c=1748
d=np.array(de[c])
p=np.array(pr[c])
len(np.unique(d)), len(np.unique(p))### it works!!
```

```python
for i in tqdm(range(len(trees.index))):
    if i==0:
        k=0
    else:
        k=trees.index[i-1]
    len(np.unique(halos.iloc[k:trees.index[i]]['id(1)']))-(trees.index[i]-k)
```
